Para generar fechas aleatorias entre rangos utilizamos la función del siguiente recurso:

https://es.stackoverflow.com/questions/418863/como-generar-una-fecha-aleatoria-entre-dos-rangos-con-random-randit-en-python

```python
from random import randint
from datetime import datetime


DATE_FORMAT = '%d/%m/%Y'


def get_random_date_between(date_a, date_b):    
    datetime_object_a = datetime.strptime(date_a, DATE_FORMAT)
    datetime_object_b = datetime.strptime(date_b, DATE_FORMAT)

    random_ordinal = randint(
        datetime_object_a.toordinal(),
        datetime_object_b.toordinal(),
    )

    random_date = datetime.fromordinal(random_ordinal)
    formatted_date_string = random_date.strftime(DATE_FORMAT)

    return formatted_date_string

get_random_date_between('17/5/2019', '30/6/2022')   
```

```python
# Generamos fechas aleatorias de registro de voluntarios

f_insc = {'Fecha inscripcion': []}
for i in np.arange(0,500):
    rand = get_random_date_between('17/5/2019', '30/6/2022')
    f_insc['Fecha inscripcion'].append(rand)

inscripcion = pd.DataFrame(f_insc)
```

En la tabla de voluntarios añadiremos una columna con la media de las puntuaciones indicadas en la tabla de opiniones

```python
star = pd.Series(opinions.groupby(['Nombre y Apellidos']).Estrellas.mean().round(1), name='Estrellas')
count = pd.Series(opinions.groupby(['Nombre y Apellidos']).Opinion.count(), name='N_opinion')
df3 = pd.DataFrame([star, count])
```
